```
import seaborn as sns; sns.set

iris = sns.load_dataset('iris')

iris.shape
```

```
sns.pairplot(iris,hue='species',corner=True,height=1.2)
```

#separer features

```
#isoler features
#features,cible
#target
y = iris.species.astype('category').cat.codes
vocabulaire = iris.species.astype('category').cat.categories
x = iris.drop('species',axis='columns')
x.shape,y.shape
```

#sparer data entre training set et test set

```
from sklearn.model_selection import train_test_split

splits = train_test_split(x,y, train_size=0.55,stratify=y,random_state=100) #separe le dataset en deux parties 1 pour entrainement a la prediction et l'autre pour le test operationnel
                                              # stratify important car permet de rcolter des reponses a comparer auxprediction de manire a tester le plus de cas possibles ( empeche de prendre ici que 1,2 ou 3 en rponse)

len(splits)
for each in splits:
  #print (each)
  print(each.shape)

x_train = splits[0]
x_test = splits[1]
y_train = splits[2]
y_test = splits[3]
```

#entrainer

```
from sklearn.neighbors import KNeighborsClassifier as KNN
knn = KNN(n_neighbors=3)

knn.fit(x_train,y_train)
```

#tester

```
def accuracy(preds, target):
  M = target.shape[0] #nb d'ex
  total_correct = (preds == target).sum()
  accuracy = (total_correct/M)
  return accuracy
```

```
pred = knn.predict(x_test)
pred
accuracy(pred,y_test)
```
